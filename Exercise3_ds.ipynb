```python
#3
import matplotlib.pyplot as plt
with open("test.txt") as f:
    data = f.read()
data = data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
x = list(map(int,x))
y = list(map(int,y))
plt.plot(x, y)
print(x)
print(y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample graph!')
plt.show()
```

```python
#6
#(i)
import numpy as np
import matplotlib.pyplot as plt
data = { 'Java':22.2,
        'Python':17.6,'PHP':8.8,'JavaScript':8,'C#':77, 'C++':6.7}
courses = list(data.keys())
values = list(data.values())
plt.bar(courses, values, color ='green',width=0.4)
 
plt.xlabel("Programming languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.show()
```

```python
#(iii)
import numpy as np
import matplotlib.pyplot as plt
data = { 'Java':22.2,
        'Python':17.6,'PHP':8.8,'JavaScript':8,'C#':77, 'C++':6.7}
courses = list(data.keys())
values = list(data.values())
plt.bar(courses, values,width=0.4,color=['black', 'red', 'green', 'blue', 'cyan','yellow'])
plt.xlabel("Programming languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.show()
```

```python
#7
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['G1','G2','G3','G4','G5']
Men= np.array([22, 30, 35, 35, 26])
Women = np.array([25, 32, 30, 35, 29])
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Men, 0.4, label = 'Men',color="green")
plt.bar(X_axis + 0.2, Women, 0.4, label = 'Women',color="red")
  
plt.xticks(X_axis, X)
plt.xlabel("Scores")
plt.ylabel("Person")
plt.title("Scores by Person")
plt.legend()
plt.show()
```

```python
#9
import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('medal.csv')
country_data = df["country"]
medal_data = df["gold_medal"]
colors = ['black', 'red', 'green', 'blue', 'cyan','yellow']
plt.pie(medal_data, labels=country_data,colors=colors)
plt.title("Gold medal achievements of five most successful\n"+"countries in 2016 Summer Olympics")
plt.show()
```
