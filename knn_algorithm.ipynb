```python
#Loading dataset iris
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
fname=iris.feature_names
print("------Feature variables  in dataset------")
print(fname)
tname=iris.target_names
print("------Target variables in dataset------")
print(tname)
print("------Size of data------")
print(x.shape)
print("------First 10 datas------")
print(x[:10])
```

```python
#Splitting into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3)
print("----Training data's size----")
print(x_train.shape)
print("----Training data----")
print(x_train)
print("----Testing data's size----")
print(x_test.shape)
print("----Testing data----")
print(x_test)
```

```python
#Train or fit the model
from sklearn.neighbors import KNeighborsClassifier
c_knn=KNeighborsClassifier(n_neighbors=5)
c_knn.fit(x_train,y_train)
y_pred=c_knn.predict(x_test)
print("----Predicted target values----")
print(y_pred)
print("-----Original target values----")
print(y_test)
```

```python
#Accuracy
from sklearn import metrics
print("-----Accuracy of predicted and original target values-----\n",metrics.accuracy_score(y_test,y_pred))
```

```python
#Prediction using model
sample=[[4.2,5.6,6.4,7.3]]
pred=c_knn.predict(sample)
pred_v=[iris.target_names[p] for p in pred]
print("----Predicted target name----")
print(pred_v)
```
